(2)在两总体的协方差矩阵相等和不等的情况下，进行Bayes 判别分析，并计算判别率的似然估计与交叉确认估计

```python
# 先验概率
n1 = np.sum(y_train == 1)
n2 = np.sum(y_train == 2)
p1 = n1 / (n1 + n2)
p2 = n2 / (n1 + n2)

# 协方差矩阵相等时的 Bayes 判别
X1 = X_train[y_train == 1]
X2 = X_train[y_train == 2]
cov1 = np.cov(X1, rowvar=False, bias=False)
cov2 = np.cov(X2, rowvar=False, bias=False)
Σ_pooled = ((n1 - 1) * cov1 + (n2 - 1) * cov2) / (n1 + n2 - 2)
Σ_pooled_inv = np.linalg.inv(Σ_pooled)
log_prior_ratio = np.log(p2 / p1)

m1 = X1.mean(axis=0)
m2 = X2.mean(axis=0)
alpha = Σ_pooled_inv.dot(m2 - m1)          
beta = (
    -0.5 * (m2 @ Σ_pooled_inv @ m2)
    +0.5 * (m1 @ Σ_pooled_inv @ m1)
    + log_prior_ratio
)

def bayes_equal_cov(x):
    return 2 if (alpha @ x + beta) > 0 else 1


y_pred_equal = np.array([bayes_equal_cov(x) for x in X_train])
err_equal_likelihood = np.mean(y_pred_equal != y_train)

loo_err_equal = 0
for idx in train_idx:
    idx_train = train_idx[train_idx != idx]
    X_loo = features[idx_train]
    y_loo = labels[idx_train]
    n1_loo = np.sum(y_loo == 1)
    n2_loo = np.sum(y_loo == 2)
    p1_loo = n1_loo / (n1_loo + n2_loo)
    p2_loo = n2_loo / (n1_loo + n2_loo)
    X1_loo = X_loo[y_loo == 1]
    X2_loo = X_loo[y_loo == 2]
    cov1_loo = np.cov(X1_loo, rowvar=False, bias=False)
    cov2_loo = np.cov(X2_loo, rowvar=False, bias=False)
    Σp_loo = ((n1_loo - 1)*cov1_loo + (n2_loo - 1)*cov2_loo) / (n1_loo + n2_loo - 2)
    Σp_loo_inv = np.linalg.inv(Σp_loo)
    log_pr_loo = np.log(p2_loo / p1_loo)
    m1_loo = X1_loo.mean(axis=0)
    m2_loo = X2_loo.mean(axis=0)
    alpha_loo = Σp_loo_inv.dot(m2_loo - m1_loo)
    beta_loo = (
        -0.5 * (m2_loo @ Σp_loo_inv @ m2_loo)
        +0.5 * (m1_loo @ Σp_loo_inv @ m1_loo)
        + log_pr_loo
    )
    x_test = features[idx]
    true_label = labels[idx]
    pred_label = 2 if (alpha_loo @ x_test + beta_loo) > 0 else 1
    if pred_label != true_label:
        loo_err_equal += 1

err_equal_cv = loo_err_equal / len(train_idx)

# 协方差矩阵不等时的 Bayes 判别（QDA）

Σ1 = np.cov(X1, rowvar=False, bias=False)
Σ2 = np.cov(X2, rowvar=False, bias=False)
Σ1_inv = np.linalg.inv(Σ1)
Σ2_inv = np.linalg.inv(Σ2)
log_det_Σ1 = np.log(np.linalg.det(Σ1))
log_det_Σ2 = np.log(np.linalg.det(Σ2))

def bayes_unequal_cov(x):
    d1 = -0.5 * ((x - m1) @ Σ1_inv @ (x - m1).T) - 0.5 * log_det_Σ1 + np.log(p1)
    d2 = -0.5 * ((x - m2) @ Σ2_inv @ (x - m2).T) - 0.5 * log_det_Σ2 + np.log(p2)
    return 1 if d1 > d2 else 2

y_pred_uneq = np.array([bayes_unequal_cov(x) for x in X_train])
err_uneq_likelihood = np.mean(y_pred_uneq != y_train)
loo_err_uneq = 0
for idx in train_idx:
    idx_train = train_idx[train_idx != idx]
    X_loo = features[idx_train]
    y_loo = labels[idx_train]
    n1_loo = np.sum(y_loo == 1)
    n2_loo = np.sum(y_loo == 2)
    p1_loo = n1_loo / (n1_loo + n2_loo)
    p2_loo = n2_loo / (n1_loo + n2_loo)
    X1_loo = X_loo[y_loo == 1]
    X2_loo = X_loo[y_loo == 2]
    m1_loo = X1_loo.mean(axis=0)
    m2_loo = X2_loo.mean(axis=0)
    Σ1_loo = np.cov(X1_loo, rowvar=False, bias=False)
    Σ2_loo = np.cov(X2_loo, rowvar=False, bias=False)
    Σ1_loo_inv = np.linalg.inv(Σ1_loo)
    Σ2_loo_inv = np.linalg.inv(Σ2_loo)
    log_det1_loo = np.log(np.linalg.det(Σ1_loo))
    log_det2_loo = np.log(np.linalg.det(Σ2_loo))
    x_test = features[idx]
    true_label = labels[idx]
    # δ1, δ2 留一法
    d1_loo = -0.5 * ((x_test - m1_loo) @ Σ1_loo_inv @ (x_test - m1_loo).T) \
             - 0.5 * log_det1_loo + np.log(p1_loo)
    d2_loo = -0.5 * ((x_test - m2_loo) @ Σ2_loo_inv @ (x_test - m2_loo).T) \
             - 0.5 * log_det2_loo + np.log(p2_loo)
    pred_label = 1 if d1_loo > d2_loo else 2
    if pred_label != true_label:
        loo_err_uneq += 1

err_uneq_cv = loo_err_uneq / len(train_idx)

print("=== 协方差相等时 (LDA) ===")
print("训练集似然估计误判率：", err_equal_likelihood)
print("留一法交叉验证误判率：", err_equal_cv)

print("\n=== 协方差不等时 (QDA) ===")
print("训练集似然估计误判率：", err_uneq_likelihood)
print("留一法交叉验证误判率：", err_uneq_cv)
```

```python
X_full = df[['x1', 'x2', 'x3', 'x4']].values  
y = df['品种'].values                         
N = len(y)

def eval_qda_lda(idxs):
    X = X_full[:, idxs]
    idx1 = np.where(y == 1)[0]
    idx2 = np.where(y == 2)[0]
    m1 = X[idx1].mean(axis=0);  S1 = np.cov(X[idx1], rowvar=False)
    m2 = X[idx2].mean(axis=0);  S2 = np.cov(X[idx2], rowvar=False)
    m3 = X[idx3].mean(axis=0);  S3 = np.cov(X[idx3], rowvar=False)
    S1_inv, S2_inv, S3_inv = np.linalg.inv(S1), np.linalg.inv(S2), np.linalg.inv(S3)
    ld1, ld2, ld3 = np.log(np.linalg.det(S1)), np.log(np.linalg.det(S2)), np.log(np.linalg.det(S3))
    p = 1/3 

    # QDA 判别函数
    def pred_qda(x):
        d1 = -0.5*((x-m1)@S1_inv@(x-m1).T) - 0.5*ld1 + np.log(p)
        d2 = -0.5*((x-m2)@S2_inv@(x-m2).T) - 0.5*ld2 + np.log(p)
        d3 = -0.5*((x-m3)@S3_inv@(x-m3).T) - 0.5*ld3 + np.log(p)
        return np.argmax([d1, d2, d3]) + 1

    # LDA 判别函数
    S_pooled = ((len(idx1)-1)*S1 + (len(idx2)-1)*S2 + (len(idx3)-1)*S3) / (N - 3)
    S_p_inv = np.linalg.inv(S_pooled)
    means = [m1, m2, m3]
    def pred_lda(x):
        scores = [x@S_p_inv@m - 0.5*(m@S_p_inv@m) for m in means]
        return np.argmax(scores) + 1

    # 回代估计
    y_pred_qda = np.array([pred_qda(x) for x in X])
    err_qda_app = np.mean(y_pred_qda != y)
    y_pred_lda = np.array([pred_lda(x) for x in X])
    err_lda_app = np.mean(y_pred_lda != y)

    # LOOCV 误判率
    loo_err_qda = 0
    loo_err_lda = 0
    for i in range(N):
        mask = np.arange(N) != i
        X_loo, y_loo = X[mask], y[mask]
        idx1_loo = np.where(y_loo == 1)[0]
        idx2_loo = np.where(y_loo == 2)[0]
        idx3_loo = np.where(y_loo == 3)[0]
        m1_loo = X_loo[idx1_loo].mean(axis=0);  S1_loo = np.cov(X_loo[idx1_loo], rowvar=False)
        m2_loo = X_loo[idx2_loo].mean(axis=0);  S2_loo = np.cov(X_loo[idx2_loo], rowvar=False)
        m3_loo = X_loo[idx3_loo].mean(axis=0);  S3_loo = np.cov(X_loo[idx3_loo], rowvar=False)
        S1i_loo = np.linalg.inv(S1_loo);  S2i_loo = np.linalg.inv(S2_loo);  S3i_loo = np.linalg.inv(S3_loo)
        ld1_loo = np.log(np.linalg.det(S1_loo));  ld2_loo = np.log(np.linalg.det(S2_loo));  ld3_loo = np.log(np.linalg.det(S3_loo))
        # QDA
        d1_loo = -0.5*((X[i]-m1_loo)@S1i_loo@(X[i]-m1_loo).T) - 0.5*ld1_loo + np.log(p)
        d2_loo = -0.5*((X[i]-m2_loo)@S2i_loo@(X[i]-m2_loo).T) - 0.5*ld2_loo + np.log(p)
        d3_loo = -0.5*((X[i]-m3_loo)@S3i_loo@(X[i]-m3_loo).T) - 0.5*ld3_loo + np.log(p)
        pred_q = np.argmax([d1_loo, d2_loo, d3_loo]) + 1
        if pred_q != y[i]:
            loo_err_qda += 1
        # LDA
        S_p_loo = ((len(idx1_loo)-1)*S1_loo + (len(idx2_loo)-1)*S2_loo + (len(idx3_loo)-1)*S3_loo) / (N - 1 - 3)
        S_p_loo_inv = np.linalg.inv(S_p_loo)
        m_loos = [m1_loo, m2_loo, m3_loo]
        scores_loo = [X[i]@S_p_loo_inv@m - 0.5*(m@S_p_loo_inv@m) for m in m_loos]
        pred_l = np.argmax(scores_loo) + 1
        if pred_l != y[i]:
            loo_err_lda += 1

    err_qda_cv = loo_err_qda / N
    err_lda_cv = loo_err_lda / N

    return {
        'err_qda_app': err_qda_app, 'err_qda_cv': err_qda_cv,
        'err_lda_app': err_lda_app, 'err_lda_cv': err_lda_cv
    }

combos = {
    'X1,X2':   [0, 1],
    'X2,X4':   [1, 3],
    'X1,X3,X4':[0, 2, 3],
    'X1,X2,X3,X4':[0, 1, 2, 3]
}

results = {name: eval_qda_lda(idxs) for name, idxs in combos.items()}

for name, res in results.items():
    print(f"{name}：")
    print(f"  QDA回代误判率 = {res['err_qda_app']:.4f}, QDA LOO-CV 误判率   = {res['err_qda_cv']:.4f}")
    print(f"  LDA 回代误判率  = {res['err_lda_app']:.4f}, LDA LOO-CV 误判率 = {res['err_lda_cv']:.4f}")
    print()
```
