## 6.6

欧洲各国语言有许多相似之处，有的甚至十分相近。以  
- E 表示英语 (English)  
- N 表示挪威语 (Norwegian)  
- Da 表示丹麦语 (Danish)  
- Du 表示荷兰语 (Dutch)  
- G 表示德语 (German)  
- Fr 表示法语 (French)  
- S 表示西班牙语 (Spanish)  
- I 表示意大利语 (Italian)  
- P 表示波兰语 (Polish)  
- H 表示匈牙利语 (Hungarian)  
- Fi 表示芬兰语 (Finnish)  

这 11 种语言中，人们以两种语言对 “1 ~ 10” 数字拼写中第一个字母不相同的个数，定义两种语言间的“距离”。这种“距离”即广义距离。例如，英语（English）和挪威语（Norwegian）仅有数字 “1” 与 “8” 的首字母不同，故这两种语言的距离定义为 2。

这样就得到这 11 种语言的距离矩阵 $D$（对称矩阵）：



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
from scipy.spatial.distance import squareform

# 行列顺序： ['E', 'N', 'Da', 'Du', 'G', 'Fr', 'S', 'I', 'P', 'H', 'Fi']
D_full = np.array([
    [ 0,  2,  2,  7,  6,  6,  6,  6,  7,  9,  9],
    [ 2,  0,  1,  5,  4,  6,  6,  6,  7,  8,  9],
    [ 2,  1,  0,  6,  5,  6,  5,  5,  6,  8,  9],
    [ 7,  5,  6,  0,  5,  9,  9,  9, 10,  9,  9],
    [ 6,  4,  5,  5,  0,  7,  7,  7,  8, 10,  9],
    [ 6,  6,  6,  9,  7,  0,  2,  1,  5, 10,  9],
    [ 6,  6,  5,  9,  7,  2,  0,  1,  3, 10,  9],
    [ 6,  6,  5,  9,  7,  1,  1,  0,  4, 10,  9],
    [ 7,  7,  6, 10,  8,  5,  3,  4,  0, 10,  9],
    [ 9,  8,  8,  9, 10, 10, 10, 10, 10,  0,  8],
    [ 9,  9,  9,  9,  9,  9,  9,  9,  9,  8,  0]
])

labels = ['E','N','Da','Du','G','Fr','S','I','P','H','Fi']

D_condensed = squareform(D_full)

Z_single   = hierarchy.linkage(D_condensed, method='single')
Z_complete = hierarchy.linkage(D_condensed, method='complete')
Z_average  = hierarchy.linkage(D_condensed, method='average')
Z_centroid = hierarchy.linkage(D_condensed, method='centroid')

print("最短距离法")
print(Z_single)
print("\n最长距离法")
print(Z_complete)
print("\n类平均距离法")
print(Z_average)
print("\n重心距离法")
print(Z_centroid)

plt.figure(figsize=(12, 10))

# 最短距离法
plt.subplot(2,2,1)
dn1 = hierarchy.dendrogram(
    Z_single,
    labels=labels,
    orientation='top',
    leaf_rotation=90
)
plt.title("最短距离法")

# 最长距离法
plt.subplot(2,2,2)
dn2 = hierarchy.dendrogram(
    Z_complete,
    labels=labels,
    orientation='top',
    leaf_rotation=90
)
plt.title("最长距离法")

# 类平均距离法
plt.subplot(2,2,3)
dn3 = hierarchy.dendrogram(
    Z_average,
    labels=labels,
    orientation='top',
    leaf_rotation=90
)
plt.title("类平均距离法")

# 重心距离法
plt.subplot(2,2,4)
dn4 = hierarchy.dendrogram(
    Z_centroid,
    labels=labels,
    orientation='top',
    leaf_rotation=90
)
plt.title("重心距离法")

plt.tight_layout()
plt.show()
```

## 6.8 

对 1975 年 1 月至 1976 年 12 月纽约证券交易所的 5 种股票（看作 5 个变量）的周反弹率的连续 100 周观测数据，求得变量相关系数矩阵为

$
R \;=\;
\begin{bmatrix}
1.000 & 0.577 & 0.509 & 0.387 & 0.462 \\
0.577 & 1.000 & 0.599 & 0.389 & 0.322 \\
0.509 & 0.599 & 1.000 & 0.436 & 0.426 \\
0.387 & 0.389 & 0.436 & 1.000 & 0.523 \\
0.462 & 0.322 & 0.426 & 0.523 & 1.000 
\end{bmatrix}
$

用适当变换将$R$变为距离矩阵，作如下分析：

(1)用最短距离法（Single Linkage）进行变量聚类，并画出谱系图（Dendrogram）；  
(2)用最长距离法（Complete Linkage）进行变量聚类，并画出谱系图；  
(3)用类平均距离法（Average Linkage）进行变量聚类，并画出谱系图。



```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import squareform

R = np.array([
 [1.000, 0.577, 0.509, 0.387, 0.462],
 [0.577, 1.000, 0.599, 0.389, 0.322],
 [0.509, 0.599, 1.000, 0.436, 0.426],
 [0.387, 0.389, 0.436, 1.000, 0.523],
 [0.462, 0.322, 0.426, 0.523, 1.000]
])

D_full = 1 - R

d_condensed = squareform(D_full)

labels = ['X1', 'X2', 'X3', 'X4', 'X5']

Z_single   = linkage(d_condensed, method='single')   # 最短距离法
Z_complete = linkage(d_condensed, method='complete') # 最长距离法
Z_average  = linkage(d_condensed, method='average')  # 类平均距离法

print("最短距离法")
print(Z_single[:2])
print("\n最长距离法")
print(Z_complete[:2])
print("\n类平均距离法")
print(Z_average)

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
dendrogram(
 Z_single,
 labels=labels,
 orientation='top',
 leaf_rotation=90
)
plt.title("最短距离法")
plt.ylabel("Distance")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.subplot(1, 3, 2)
dendrogram(
 Z_complete,
 labels=labels,
 orientation='top',
 leaf_rotation=90
)
plt.title("最长距离法")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.subplot(1, 3, 3)
dendrogram(
 Z_average,
 labels=labels,
 orientation='top',
 leaf_rotation=90
)
plt.title("类平均距离法")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
