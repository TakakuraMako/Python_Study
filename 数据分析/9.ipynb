## 7.3
设一卷磁带上的缺陷数服从 Poisson 分布 $P(\theta)$，其中 $\theta$ 可取1.0和1.5中的一个。
设 $\theta$ 的先验分布为：$\pi(1.0) = 0.4, \quad \pi(1.5) = 0.6$若一卷磁带发现了3个缺陷，求 $\theta$ 的后验分布。

设 $X \sim \text{Poisson}(\theta)$，观测值为 $x = 3$。根据贝叶斯公式：
$$
\pi(\theta \mid x) = \frac{\pi(\theta) \cdot P(x \mid \theta)}{P(x)}
$$

计算似然函数
$$
P(x = 3 \mid \theta) = \frac{e^{-\theta} \cdot \theta^3}{3!} = \frac{e^{-\theta} \cdot \theta^3}{6}
$$

当 $\theta = 1.0$ 时：
$$
P(3 \mid 1.0) = \frac{e^{-1.0} \cdot 1.0^3}{6} = \frac{e^{-1.0}}{6} \approx \frac{0.3679}{6} \approx 0.0613
$$

当 $\theta = 1.5$ 时：
$$
P(3 \mid 1.5) = \frac{e^{-1.5} \cdot 1.5^3}{6} = \frac{e^{-1.5} \cdot 3.375}{6} \approx \frac{0.2231 \cdot 3.375}{6} \approx \frac{0.752}{6} \approx 0.1253
$$

边际似然概率 $P(3)$
$$
P(3) = \pi(1.0) \cdot P(3 \mid 1.0) + \pi(1.5) \cdot P(3 \mid 1.5)
$$
$$
= 0.4 \cdot 0.0613 + 0.6 \cdot 0.1253 = 0.0245 + 0.0752 = 0.0997
$$
计算后验分布

$$
\pi(1.0 \mid 3) = \frac{0.4 \cdot 0.0613}{0.0997} \approx \frac{0.0245}{0.0997} \approx 0.246
$$
$$
\pi(1.5 \mid 3) = \frac{0.6 \cdot 0.1253}{0.0997} \approx \frac{0.0752}{0.0997} \approx 0.754
$$

最终答案
$$
\boxed{
\pi(1.0 \mid 3) \approx 0.246, \quad \pi(1.5 \mid 3) \approx 0.754
}
$$

## 7.4

设随机变量 $X$ 的概率密度是
$$
f(x \mid \theta) =
\begin{cases}
\frac{2x}{\theta^2}, & 0 < x < \theta \\
0, & \text{其他}
\end{cases}
$$
其中 $0 < \theta < 1$，设 $\theta$ 的先验分布是均匀分布 $U(0,1)$，求 $\theta$ 的后验分布。


根据贝叶斯公式有：
$$
\pi(\theta \mid x) \propto \pi(\theta) \cdot f(x \mid \theta)
$$

先验分布
由于 $\theta \sim U(0,1)$，故其先验概率密度为：
$$
\pi(\theta) = 1, \quad 0 < \theta < 1
$$

似然函数
在观测到 $x$ 后，视 $x$ 为常量，似然函数为：
$$
f(x \mid \theta) =
\begin{cases}
\displaystyle \frac{2x}{\theta^2}, & x < \theta < 1 \\
0, & \text{其他}
\end{cases}
$$

后验分布（非标准化）
将先验与似然相乘，得后验分布（未归一化）：
$$
\pi(\theta \mid x) \propto
\begin{cases}
\displaystyle \frac{2x}{\theta^2}, & x < \theta < 1 \\
0, & \text{其他}
\end{cases}
$$

归一化常数
为了得到标准化的后验分布，需对未标准化分布在 $(x, 1)$ 上积分：
$$
\int_x^1 \frac{2x}{\theta^2} \, d\theta
= 2x \int_x^1 \frac{1}{\theta^2} \, d\theta
= 2x \left[ -\frac{1}{\theta} \right]_x^1
= 2x \left( -1 + \frac{1}{x} \right)
= 2(1 - x)
$$

后验分布（标准化）
将归一化因子代入，得标准化后的后验分布：
$$
\pi(\theta \mid x) =
\begin{cases}
\displaystyle \frac{1}{2(1 - x)} \cdot \frac{2x}{\theta^2}
= \frac{2x}{(1 - x)\theta^2}, & x < \theta < 1 \\
0, & \text{其他}
\end{cases}
$$

最终答案
$$
\boxed{
\pi(\theta \mid x) =
\begin{cases}
\displaystyle \frac{2x}{(1 - x)\theta^2}, & x < \theta < 1 \\
0, & \text{其他}
\end{cases}
}
$$




## 7.7
设一批产品的不合格品率为 $\theta$，检查是一个接一个地进行，直到发现第一个不合格品就停止检查。若 $x$ 为发现第一个不合格品时已检查的产品数，则 $x$ 服从几何分布：
$$
f(x \mid \theta) = \theta (1 - \theta)^{x - 1}, \quad x = 1, 2, \dots
$$

若 $\theta$ 只取 $\frac{1}{4}$、$\frac{1}{2}$、$\frac{3}{4}$ 这三个值，且取这三个值的概率相同。今获得一个样本观测值 $x = 3$，求：
(1) $\theta$ 的后验分布；
(2) $\theta$ 的最大后验估计 $\hat{\theta}_{\mathrm{MD}}$。

计算似然函数

给定 $x = 3$，有：
$$
f(3 \mid \theta) = \theta (1 - \theta)^2
$$
对三个可能的 $\theta$ 值：
$$
\begin{aligned}
\theta = \frac{1}{4} &\Rightarrow f = \frac{1}{4} \cdot \left(\frac{3}{4}\right)^2 = \frac{9}{64} \\
\theta = \frac{1}{2} &\Rightarrow f = \frac{1}{2} \cdot \left(\frac{1}{2}\right)^2 = \frac{1}{8} = \frac{8}{64} \\
\theta = \frac{3}{4} &\Rightarrow f = \frac{3}{4} \cdot \left(\frac{1}{4}\right)^2 = \frac{3}{64}
\end{aligned}
$$

计算后验分布

后验分布正比于先验乘以似然。因为先验相同为 $\frac{1}{3}$，只需归一化似然即可。

归一化系数为：

$$
\frac{9 + 8 + 3}{64} = \frac{20}{64}
$$

所以后验概率为：
$$
\begin{aligned}
\pi\left(\frac{1}{4} \mid x=3\right) &= \frac{9}{20} \\
\pi\left(\frac{1}{2} \mid x=3\right) &= \frac{8}{20} = \frac{2}{5} \\
\pi\left(\frac{3}{4} \mid x=3\right) &= \frac{3}{20}
\end{aligned}
$$

最大后验估计

后验概率最大的为 $\theta = \frac{1}{4}$，因此：
$$
\hat{\theta}_{\mathrm{MD}} = \frac{1}{4}
$$

最终答案

$$
\boxed{
\begin{aligned}
&\pi\left(\frac{1}{4} \mid x = 3\right) = \frac{9}{20}, \quad
\pi\left(\frac{1}{2} \mid x = 3\right) = \frac{2}{5}, \quad
\pi\left(\frac{3}{4} \mid x = 3\right) = \frac{3}{20} \\
&\hat{\theta}_{\mathrm{MD}} = \frac{1}{4}
\end{aligned}
}
$$


## 7.13
某产品的长度服从正态分布 $N(\mu, \sigma^2)$（$\mu$ 已知），设 $\theta = \sigma^2$ 的先验分布为：
$$
\pi(\theta) \propto \frac{1}{\theta}
$$

取 200 个产品做试验，得到误差平方和：$t = \sum_{i=1}^{200} (x_i - \mu)^2 = 30.56$

求 $\theta$ 的 0.95 等尾可信区间。

由于 $\mu$ 已知，且 $X_i \sim N(\mu, \sigma^2)$，设 $\theta = \sigma^2$，先验为非信息先验 $\pi(\theta) \propto \frac{1}{\theta}$。

根据贝叶斯推断，有：

$$
\theta \mid x_1,\dots,x_n \sim \text{Inverse-Gamma}\left(\alpha = \frac{n}{2}, \beta = \frac{t}{2} \right)
$$

本题中：

$$
n = 200, \quad t = 30.56 \Rightarrow \theta \mid \text{data} \sim \text{Inverse-Gamma}(100, 15.28)
$$
等价地，设 $Y = \frac{1}{\theta}$，则：

$$
Y \sim \text{Gamma}(100, \text{rate} = \frac{1}{15.28})
$$

要求满足：

$$
P(\theta_L \leq \theta \leq \theta_U) = 0.95
$$

等价于：

$$
P\left( \frac{1}{\theta_U} \leq Y \leq \frac{1}{\theta_L} \right) = 0.95
$$

Gamma 分布的分位数

$$
\theta_L \approx \frac{1}{y_{0.975}}, \quad \theta_U \approx \frac{1}{y_{0.025}}
$$

最终得到：

$$
\boxed{
\theta \in [0.132, \; 0.202] \quad \text{为 0.95 等尾可信区间}
}
$$

## 7.15

设一批产品的废品率为 $\theta$，其先验分布 $\pi(\theta) \sim U(0,1)$。从该批产品中有放回地抽取容量为 600 的样本，记其废品数量为 $x$。作下列 Bayes 假设检验：

$$
H_0: \theta \leq 0.03 \quad \text{vs.} \quad H_1: \theta > 0.03
$$

试制定一个抽样方案，说明何时接受 $H_0$，何时拒绝 $H_0$。



由于 $x \sim \text{Binomial}(n=600, \theta)$，且先验为 $U(0,1) \sim \text{Beta}(1,1)$，后验分布为：

$$
\theta \mid x \sim \text{Beta}(x + 1, 601 - x)
$$

根据 Bayes 决策准则：

若 $P(\theta \leq 0.03 \mid x) > 0.95$，则接受 $H_0$；

若 $P(\theta \leq 0.03 \mid x) < 0.05$，则拒绝 $H_0$；

否则，不作决定或继续抽样。


使用 Beta 分布的累积分布函数 $\mathrm{BetaCDF}(0.03; x+1, 601 - x)$ 计算后验概率。数值计算得：

$$
\begin{aligned}
&x \leq 9 \Rightarrow P(\theta \leq 0.03 \mid x) > 0.95 \Rightarrow \text{接受 } H_0 \\
&x \geq 17 \Rightarrow P(\theta \leq 0.03 \mid x) < 0.05 \Rightarrow \text{拒绝 } H_0
\end{aligned}
$$


制定如下抽样判定规则：

若观察到废品数 $x \leq 9$，接受 $H_0$；

若 $x \geq 17$，拒绝 $H_0$；

若 $10 \leq x \leq 16$，保持中立或补抽样。
