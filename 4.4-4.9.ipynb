（2）从R出发做主成分分析，求各主成分的贡献率及前两个主成分的累计贡献率

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 标准化数据
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data4_5.select_dtypes(include=[np.number]))
# 主成分分析
pca = PCA(n_components=8)
# 输出特征值
pca.fit(data_scaled)
eigenvalues = pca.explained_variance_
print("主成分分析结果：", eigenvalues)

# 计算各主成分贡献率
explained_variance_ratio = pca.explained_variance_ratio_
print("各主成分贡献率：", explained_variance_ratio)

# 前两个主成分累计贡献率
cumulative_variance = np.cumsum(explained_variance_ratio)
print("前两个主成分累计贡献率：", cumulative_variance[:2])
```

4.9
从样本协方差矩阵 $\Sigma$ 和样本相关系数矩阵R出发做典型相关分析，求各典型变量及典型相关系数，检验各典型变量对是否显著相关 ($\alpha = 0.05$)。比较两种情况下的结果的异同。

```python
from sklearn.cross_decomposition import CCA
from scipy.stats import chi2

# 原始数据
data4_9 = {
    "x1": [
        191, 195, 181, 183, 176, 208, 189, 197, 188, 192,
        179, 183, 174, 190, 188, 163, 195, 186, 181, 175,
        192, 174, 176, 197, 190
    ],
    "x2": [
        155, 149, 148, 153, 144, 157, 150, 159, 152, 150,
        158, 147, 150, 159, 151, 137, 155, 153, 145, 140,
        154, 143, 139, 167, 163
    ],
    "y1": [
        179, 201, 185, 188, 171, 192, 190, 189, 197, 187,
        186, 174, 185, 195, 187, 161, 183, 173, 182, 165,
        185, 178, 176, 200, 187
    ],
    "y2": [
        145, 152, 149, 149, 142, 152, 149, 152, 159, 151,
        148, 147, 152, 157, 158, 130, 158, 148, 146, 137,
        152, 147, 143, 158, 150
    ]
}

# 构造 DataFrame 并拆分 X, Y
df = pd.DataFrame(data4_9)
X = df[['x1', 'x2']].values
Y = df[['y1', 'y2']].values
n = X.shape[0]
p, q = X.shape[1], Y.shape[1]

def cca_analysis(X, Y, scale):
    """
    对 X, Y 做典型相关分析
    scale=False 时基于样本协方差矩阵
    scale=True 时基于样本相关系数矩阵
    """
    cca = CCA(n_components=min(p, q), scale=scale)
    cca.fit(X, Y)
    U, V = cca.transform(X, Y)
    # 计算各典型变量对之间的相关系数
    canonical_corrs = [np.corrcoef(U[:, i], V[:, i])[0, 1] 
for i in range(min(p, q))]
    # Wilks' Lambda 检验
    results = []
    for i in range(len(canonical_corrs)):
        # 从 i 到最后一对的 (1 - rho^2) 连乘积
        lambda_val = np.prod([1 - canonical_corrs[j]**2 
for j in range(i, len(canonical_corrs))])
        df_chi = (p - i) * (q - i)
        stat = - (n - 1 - (p + q + 1) / 2) * np.log(lambda_val)
        p_value = chi2.sf(stat, df_chi)
        results.append({
            'canonical_pair': i + 1,
            'corr': canonical_corrs[i],
            'Wilks_lambda': lambda_val,
            'chi2_stat': stat,
            'df': df_chi,
            'p_value': p_value
        })
    return cca, canonical_corrs, results

# 基于协方差矩阵的 CCA
cca_cov, r_cov, res_cov = cca_analysis(X, Y, scale=False)
# 基于相关系数矩阵的 CCA
cca_cor, r_cor, res_cor = cca_analysis(X, Y, scale=True)

print("----- 基于协方差矩阵的 CCA 结果 -----")
print("典型相关系数：", np.round(r_cov, 6))
print("X 端典型系数矩阵：\n", np.round(cca_cov.x_weights_, 6))
print("Y 端典型系数矩阵：\n", np.round(cca_cov.y_weights_, 6))
print("Wilks' Lambda 检验：")
for item in res_cov:
    print(item)

print("\n----- 基于相关系数矩阵的 CCA 结果 -----")
print("典型相关系数：", np.round(r_cor, 6))
print("X 端典型系数矩阵：\n", np.round(cca_cor.x_weights_, 6))
print("Y 端典型系数矩阵：\n", np.round(cca_cor.y_weights_, 6))
print("Wilks' Lambda 检验：")
for item in res_cor:
    print(item)
```

典型相关系数：无论是基于样本协方差矩阵，还是基于样本相关系数矩阵做CCA，计算得到的两对典型相关系数均为$ \rho_1 = 0.7885, \rho_2 = 0.5537 $。这表明第一对典型变量之间的相关程度约为 0.7885，第二对几乎不相关。

典型变量系数：虽然两种方法下得到的系数数值不同，但它们对应的典型变量在原始尺度上的组合含义是相同的——两者都能提取出与第一对典型相关最强的线性组合；只不过当用 $R$ 时，数据先做了均值归一化，所以权重向量的方向和比例与直接用 $\Sigma$ 时不同。

显著性检验：

1.样本容量 $n=25$，$p=q=2$，对于第 $i$ 对典型相关，Wilks’ λ 计算及检验统计量如下：

第一对（$i=1$）：

$\Lambda_1 \approx 0.3771629$，

$\chi^2$ 统计量 $\approx 20.9642$， $\mathrm{df}=(2-0+0)(2-0+0)=4$，

$p \approx 0.00032$。
因此拒绝“第一对典型相关为 0” 的原假设：第一对典型相关显著不为 0。

第二对（$i=2$）：

$\Lambda_2 \approx 0.9971120$，

$\chi^2$ 统计量 $\approx 0.06218$， $\mathrm{df}=(2-1)(2-1)=1$，

$p \approx 0.80308$。
因此无法拒绝“第二对典型相关为 0” 的原假设：第二对典型变量之间不显著相关。

$\quad$上述 Wilks' λ 测试结果在两种做法下完全一致—：仅第一对典型变量显著，第二对不显著。

2.两种情况下的异同点：基于 $\Sigma$ 得到的权重体现的是“原始单位下”对典型变量的线性组合；基于 $R$ 得到的权重体现的是“标准化单位下”对典型变量的线性组合；但它们提取出的典型变量对在说明两组变量的最大相关方向上是一致的。
